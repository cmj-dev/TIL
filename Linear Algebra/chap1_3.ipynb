## Basic of numpy vectors

### Vector copying

```python
import numpy as np

a = np.array([1,2,3,4])
b = a.copy() # copy by value
c = a # copy by reference
a[2] = 10
a, b, c
```

### Vector stacking

```python
import numpy as np

x = np.array([1,2,3,4]); y = np.array([3,2,1])
np.concatenate((x, y))
```

### Zero vector, One vector, Unit vector

```python
# zero vector
np.zeros(4)
```

```python
# one vector
np.ones(3)
```

```python
# unit vector
unit = lambda i, n: np.concatenate((np.zeros(i), [1], np.zeros(n - i - 1)))
unit(1,10)
```

## Taylor approximation

### First-order Taylor approximation

If we gonna made First-order Taylor approximation for f(x) at point z:

$$ \hat f(x) = f(z) + \frac {f^\prime(z)} {1!}(x-z) $$

So for $ \hat f(x) $ if x is located nearby z, $ \hat f(x) $ will have a good method to predict $ f(x) $

Ex) Get the first-order taylor approximation of $f(x) = x_1 + exp(x_2 - x_1)$ at point (1, 2)

A) $\hat f(x) = f^\prime (z)(x-z) + f(z)$

So we have to calculate $f^\prime (x)$

```python
from math import exp
f = lambda x: x[0] + exp(x[1] - x[0])
grad_f = lambda x: [1-exp(x[1] - x[0]), exp(x[1] - x[0])]
z = [1,2]
# Taylor approximation
f_hat = lambda x: f(z) + np.inner(grad_f(z), (x-z))
f([1,2]), f_hat(np.array([1,2]))
```
